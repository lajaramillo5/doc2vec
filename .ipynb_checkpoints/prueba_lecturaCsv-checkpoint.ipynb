```python
import time
# importaciones básicas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
nltk.download('punkt')


# modelado de palabras
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from nltk.tokenize import word_tokenize

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.manifold import TSNE
from sklearn.linear_model import LogisticRegression

import warnings

warnings.filterwarnings("ignore",category=DeprecationWarning)
warnings.filterwarnings('ignore',message="Precision")
```

```python
#Cargar datos
lectures = pd.read_csv('all_lecture_text.csv')
lectures.head()
```

```python
#Dividir el df para entrenamiento y pruebas
train, test = train_test_split(lectures.reset_index(drop=True), test_size=.30, random_state=43)
# El modelo Doc2Vec toma 'tagged_documents' 
#tag los datos de entrenamiento
tagged_tr = [TaggedDocument(words=word_tokenize(_d.lower()),\
tags=[str(i)]) for i, _d in enumerate(train.d2v_text)]
#tag datos de prueba
tagged_test = [TaggedDocument(words=word_tokenize(_d.lower()),\
tags=[str(i)]) for i, _d in enumerate(test.d2v_text)]
```

```python
# Instanciar el modelo
model = Doc2Vec(vector_size=100, # 100 debe estar bien basado en los estándares
                window=5, #change to 8
                alpha=.025, # tasa de aprendizaje inicial                 
                min_alpha=0.00025,  # cae linealmente a esto
                min_count=2, # ignora todas las palabras con una frecuencia total inferior a esta.
                dm =1, #algorith 1=distributed memory
                workers=16)# núcleos para usar

# Construye el vocabulario en los datos de entrenamiento
model.build_vocab(tagged_tr)
```

```python

#max training epochs
max_epochs = 100

#train n epochs and save the model
t1 = time.time()
for epoch in range(max_epochs):
    print('iteration {0}'.format(epoch+1))
    model.train(tagged_tr,
                total_examples=model.corpus_count,
                epochs=model.epochs)
    # decrease the learning rate
    model.alpha -= 0.0002
    # fix the learning rate, no decay
    model.min_alpha = model.alpha
   
   

print("done!")
t2 = time.time()    
model.save("5klects1.model")
#print("Model Saved")
print("Time: {}".format(t2-t1))
```

```python
# Ahora que tenemos la incrustación entrenada, podemos usar el método de inferir vectores para convertir las oraciones de prueba en vectores 
#que se puede usar para modelar

# Extraer vectores del modelo doc2vec
X_train = np.array([model.docvecs[str(i)] for i in range(len(tagged_tr))])
y_train = train.Subject

# Extraer valores de prueba
X_test = np.array([model.infer_vector(tagged_test[i][0]) for i in range(len(tagged_test))])
y_test = test.Subject

```

```python
def heatconmat(y_true,y_pred):
    sns.set_context('talk')
    plt.figure(figsize=(9,6))
    sns.heatmap(confusion_matrix(y_true,y_pred),
                annot=True,
                fmt='d',
                cbar=False,
                cmap='gist_earth_r',
                yticklabels=sorted(y_test.unique()))
    plt.show()
    print(classification_report(y_true,y_pred))
```

```python
lrc = LogisticRegression(C=5, multi_class='multinomial', solver='saga',max_iter=1000)
lrc.fit(X_train,y_train)
y_pred = lrc.predict(X_test)
heatconmat(y_test,y_pred)
```
